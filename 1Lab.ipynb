# Задание 1.4

```python
def polar(P, a):
    return(3 * a * np.sin(P) * np.cos(P))/ (np.sin(P)**3 + np.cos(P)**3)
```

```python
def x_f(t):
    return (3 * t**2 + 1) / (3 * t**3)
```

```python
def y_f(t):
    return np.sin(t**3/3+t)
```

```python
P = np.linspace(0,2*np.pi,1000)
a = 1
r = polar(P,a)
```

```python
plt.figure(figsize=(6,6))
ax = plt.subplot(111, projection='polar')
ax.plot(P,r)
ax.set_title("Polar system")
plt.show()
```

```python
t = np.linspace(0.1, 10, 1000)
x = x_f(t)
y = y_f(t)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x,y)
plt.title("Decarte")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```
